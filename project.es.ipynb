#### Ejercicio 10. ¿Son los promedios de precio por metro cuadrado (precio/m2) de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios de precio por metro cuadrado y escribe una conclusión sobre ellos.

Pista: Crea una nueva columna llamada `pps` (*price per square* o precio por metro cuadrado) y luego analiza los valores.

```python
# TODO
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')

ds_filtered_pps = ds[ds['level5'].isin(['Valdemorillo', 'Galapagar'])]

ds['pps'] = ds_filtered_pps['price'] / ds_filtered_pps['surface']

square_meter_Valdemorillo = ds[ds['level5'] == 'Valdemorillo']['pps'].mean()
square_meter_Galapagar = ds[ds['level5'] == 'Galapagar']['pps'].mean()

print(f"El promedio del precio por metro cuadrado en Valdemorillo es de : {round(square_meter_Valdemorillo, 2)} EUR/m^2\n"
      f"El promedio del precio por metro cuadrado en Galapagar es de: {round(square_meter_Galapagar, 2)} EUR/m^2.")

diferencia_absoluta = square_meter_Galapagar - square_meter_Valdemorillo
diferencia_relativa = ((square_meter_Galapagar - square_meter_Valdemorillo) / square_meter_Valdemorillo) * 100

print(f"Esto representa una diferencia absoluta de {round(diferencia_absoluta,2)} EUR por metro cuadrado.\n"
      f"Relativamente, el precio en Galapagar es aproximadamente un {round(diferencia_relativa,2)}% más alto que en Valdemorillo.")
```

```python
# TODO: Code
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

cinturon_sur = ds_filtered[ds_filtered['level5'].isin(["Fuenlabrada","Leganés","Getafe","Alcorcón"])]

median_price = cinturon_sur.groupby('level5')['price'].median()

# Gráfico de barras 
plt.bar( x = median_price.index, height = median_price.values,
        edgecolor = 'black',
        color = 'darkblue',
        alpha = 0.9)

# Título y labels
plt.title('Mediana de los precios')
plt.xlabel('Población')
plt.ylabel('Precio')

# Textos en las barras
plt.text(-0.35,165000,median_price.loc["Alcorcón"],fontsize=12, color='w')
plt.text(0.65,140000,median_price.loc["Fuenlabrada"],fontsize=12, color='w')
plt.text(1.65,270000,median_price.loc["Getafe"],fontsize=12, color='w')
plt.text(2.65,155000,median_price.loc["Leganés"],fontsize=12, color='w')

#Forma optimizada con textos dinámicos
# value + 5000(valor cualquiera) es para que el valor no se superponga con la barra. 
for index, value in enumerate(median_price.values):
    plt.text(index, value + 5000, f'{round(value, 2)}', ha='center', fontsize=10, color='black')

plt.show()
```

```python
# TODO  
import pandas as pd
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None  # default='warn'
cinturon_sur = ds_filtered[ds_filtered['level5'].isin(["Fuenlabrada","Leganés","Getafe","Alcorcón"])]


# NUEVA COLUMNA CON LOS PRECIOS NORMALIZADOS
cinturon_sur['Normalización'] = (cinturon_sur['price'] - cinturon_sur['price'].mean()) / cinturon_sur['price'].std()

# valores normalizados de cada población
normalizacion_fuenlabrada = cinturon_sur.loc[cinturon_sur['level5'] == 'Fuenlabrada', 'Normalización']

normalizacion_leganes = cinturon_sur.loc[cinturon_sur['level5'] == 'Leganés', 'Normalización']

normalizacion_getafe = cinturon_sur.loc[cinturon_sur['level5'] == 'Getafe', 'Normalización']

normalizacion_alcorcon = cinturon_sur.loc[cinturon_sur['level5'] == 'Alcorcón', 'Normalización']

# Crear la figura y los subplots
fig, axis = plt.subplots(2, 2, figsize=(10, 8))

# Histograma Fuenlabrada
axis[0, 0].hist(normalizacion_fuenlabrada, bins=30, color='blue', edgecolor='black')
axis[0, 0].set_title('Fuenlabrada')

# Histograma Leganés
axis[0, 1].hist(normalizacion_leganes, bins=30, color='green', edgecolor='black')
axis[0, 1].set_title('Leganés')

# Histograma Getafe
axis[1, 0].hist(normalizacion_getafe, bins=30, color='red', edgecolor='black')
axis[1, 0].set_title('Getafe')

# Histograma Alcorcón
axis[1, 1].hist(normalizacion_alcorcon, bins=30, color='purple', edgecolor='black')
axis[1, 1].set_title('Alcorcón')

# Ajustar el diseño
plt.tight_layout()


plt.show()
```

```python
fig, axis = plt.subplots(figsize=(12, 6))

plt.hist(
    (normalizacion_fuenlabrada, normalizacion_leganes, normalizacion_getafe, normalizacion_alcorcon),
      bins = 20,
        #histtype='bar',
        label = ('Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'),
        color = ('blue', 'green', 'red', 'purple')
        )
plt.title('Normalized prices')
plt.legend()
plt.grid()

plt.tight_layout()

plt.show()
```

**TODO: Markdown**. En textos de estadística frecuentemente se habla de variable normalizada, también se conoce como unidad tipificada, variable centrada reducida o variable estandarizada. Normalizar una variable es simplemente expresar su magnitud en unidades de desviación estándar.
Es importante entender que normalizar una variable no cambia su valor, solo su unidad de cuenta.
Normalizar las variables nos permite comparar su distribución independientemente de su unidad de cuenta y amplitud, también nos permite sacar conclusiones sobre probabilidades y proporciones.

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

Pista: Crea una nueva columna llamada `pps` (price per square en inglés) y luego analiza los valores

```python
# TODO
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.options.mode.chained_assignment = None  # default='warn'

# Subconjunto obtenido en la pregunta 14.
cinturon_sur = ds_filtered[ds_filtered['level5'].isin(["Fuenlabrada","Leganés","Getafe","Alcorcón"])]

# Creamos una columna en el subplot cinturon_sur llamada pps (price per square). Los datos son price/surface
cinturon_sur['pps'] = cinturon_sur['price'] / cinturon_sur['surface']

# Precio por metro cuadrado de Getafe
pps_getafe = cinturon_sur[cinturon_sur['level5']=='Getafe']['pps'].mean()

# Precio por metro cuadrado de Alcorcón
pps_alcorcon = cinturon_sur[cinturon_sur['level5']=='Alcorcón']['pps'].mean()


poblaciones = ['Getafe', 'Alcorcón']
precios = [pps_getafe, pps_alcorcon]

plt.bar(poblaciones, precios, color=['blue', 'green'])
plt.ylabel('Precio Medio por m²')
plt.title('Comparación del Precio Medio por Metro Cuadrado')
plt.show()

diferencia_absoluta = pps_alcorcon - pps_getafe
diferencia_relativa = ((pps_alcorcon - pps_getafe) / pps_getafe) * 100

print(f"El precio medio del metro cuadrado en Getafe es de {round(pps_getafe,2)} EUR, mientras que en Alcorcón es de {round(pps_alcorcon,2)} EUR.\n"
      f"Esto representa una diferencia absoluta de {round(diferencia_absoluta,2)} EUR por metro cuadrado.\n"
      f"Relativamente, el precio en Alcorcón es aproximadamente un {round(diferencia_relativa,2)}% más alto que en Getafe.")

```

```python
from ipyleaflet import Map, basemaps, Marker, AwesomeIcon
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

## Aquí: traza la coordenadas de los estados
cinturon_sur = ds_filtered[ds_filtered['level5'].isin(["Fuenlabrada","Leganés","Getafe","Alcorcón"])]

latitude = cinturon_sur['latitude'].reset_index(drop=True).str.replace(',','.').astype(float) # Sustituir comas por puntos porque daba error
longitude = cinturon_sur['longitude'].reset_index(drop=True).str.replace(',','.').astype(float) # Sustituir el type por float para la correcta representación de coordenadas
level5 = cinturon_sur['level5'].reset_index(drop=True)

# Las pasamos a listas
latitude.to_list()
longitude.to_list()
level5.to_list()

# Usamos las listas para construir un dict
dict = {
    "latitude": latitude,
    "longitude": longitude,
    "level5": level5,
}

# Convertimos el dict a dataframe
dataframe = pd.DataFrame(dict)

## PON TU CÓDIGO AQUÍ:

# Función que le asigna un color a cada población
def color_by_population(level5):
    if level5 == 'Fuenlabrada':
        return 'black'
    elif level5 == 'Leganés':
        return 'red'
    elif level5 == 'Getafe':
        return 'green'
    else: 
        return 'darkblue'


mapa = Map(center=(40.32065, -3.79268), zoom=12, basemap=basemaps.OpenStreetMap.Mapnik)

# Iteración para representar el dataframe
for _, row in dataframe.iterrows():
    marker = Marker(
        location=(row['latitude'], row['longitude']),
        draggable=False,
        title=row['level5'],
        icon=AwesomeIcon(name='fa-github', marker_color=color_by_population(row['level5']))
        
    )
    mapa.add_layer(marker)

# Display the map
mapa   

```
